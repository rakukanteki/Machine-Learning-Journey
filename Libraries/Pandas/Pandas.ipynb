# Handle missing data
- use fillna to fill missing values.
- use interpolation to guess the missing value and fill it.
- use dropna to drop rows.

```python
import pandas as pd
df = pd.read_csv("Handle_weather_data.csv", parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
# Setting day as index.
df.set_index("day", inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
# Sometimes using 0 is not that we want.
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
# Now in columns the data distribution still seems a bit horrible. 
# Doing carry forward operation. We can do it horizontally or vetically. So use 'axis'.
# We can limit the forward operation. Means we can limit the steps. Use 'limit=__' 
new_df = df.fillna(method="ffill")
new_df
```

```python
# Making data frame more better.
new_df = df.interpolate(method = 'time')
new_df
```

##### Dropping

```python
# Dropping a abberant rows.
# In df, 2017-01-09 has no values.
new_df = df.dropna(how="all")
new_df
```

```python
# We keep rows that has at least one valid value.
new_df = df.dropna(thresh=1)
new_df
```

```python
# We keep rows that has at least 2 valid value.
new_df = df.dropna(thresh=2)
new_df
```

##### Going after the missing dates.

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```

# Concat Dataframes.

```python
import pandas as pd
bd_weather = pd.DataFrame({
    'city': ['dhaka','rajshahi','sylhet'],
    'temperature': [32,45,30],
    'humidity': [80,60,78]
})
bd_weather
```

```python
usa_weather = pd.DataFrame({
    'city': ['New York','Chicago','Orlando'],
    'temperature': [21,14,35],
    'humidity': [68,65,75]
})
usa_weather
```

```python
# Joining two data frames.
new_df = pd.concat([bd_weather, usa_weather])
new_df
```

```python
# For continuous index.
new_df = pd.concat([bd_weather, usa_weather], ignore_index=True)
new_df
```

```python
# Implementation of keys.
new_df = pd.concat([bd_weather, usa_weather], keys=["BD", "USA"])
new_df
```

```python
new_df.loc["USA"]
```

##### Joing two data frames by columns

```python
temp_df = pd.DataFrame({
    "city": ["dhaka","rajshahi", "sylhet"],
    "temperature": [32,45,30]
})
temp_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["dhaka","rajshahi", "sylhet"],
    "windspeed": [7,12,9]
})
windspeed_df
```

```python
new_df = pd.concat([temp_df, windspeed_df], axis=0)
new_df
```

```python
# Creating additional rows containing NaN. So removing it.
new_df = pd.concat([temp_df, windspeed_df], axis=1)
new_df
```

```python
# Say our datas are not indexed well.
temp_df = pd.DataFrame({
    "city": ["dhaka","rajshahi", "sylhet"],
    "temperature": [32,45,30]
})
temp_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["rajshahi", "dhaka"],
    "windspeed": [7,12]
})
windspeed_df
```

```python
new_df = pd.concat([temp_df, windspeed_df], axis=1)
new_df
```

```python
# But we want rows be like: dhaka  dhaka, So,
temp_df = pd.DataFrame({
    "city": ["dhaka","rajshahi", "sylhet"],
    "temperature": [32,45,30]
}, index=[0,1,2])
temp_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["rajshahi", "dhaka"],
    "windspeed": [7,12]
}, index=[1,0])
windspeed_df
```

```python
new_df = pd.concat([temp_df, windspeed_df], axis=1)
new_df
```

Now, it's working correctly.
